```python
from glob import glob
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_name=glob('D:/cuda/MonteCarloPathTracing/MonteCarloPathTracing/data/dataset/noise2/color/*.[pP][pP][mM]' )
noise_image=[]
FileList=[]
# get all file name
for x in file_name:
    FileList.append(x)
```

```python
target_width=128
target_height=128

def crop_image(img):
    crop_img=[]

    w = target_width
    h = target_height

    for j in range(6):
        for i in range(10):
        
            x=i*128
            y=j*128


            if j==5:
                y=720-128
                
            crop_img.append(img[y:y+h, x:x+w]/255)
            
    return crop_img
```

```python
def merge_img(imgArr_x,imgArr_y,imgArr_n,imgArr_d,n):
    for i in range(len(imgArr_x)):
        training_img=np.concatenate((imgArr_x[i], imgArr_y[i],imgArr_n[i],imgArr_d[i]), axis=1)
        plt.imsave(
            ("D:/training_model/20190606/mon2/"+repr(n)+"_"+repr(i)),training_img)
```

```python
path_o='D:/cuda/MonteCarloPathTracing/MonteCarloPathTracing/data/dataset/reference2/optixPathTracer'
path_n='D:/cuda/MonteCarloPathTracing/MonteCarloPathTracing/data/dataset/noise2/normal/optixPathTracer'
path_d='D:/cuda/MonteCarloPathTracing/MonteCarloPathTracing/data/dataset/noise2/depth/optixPathTracer'


for i in range( len(FileList)):
    filename_x=FileList[i]
    

    img_x=cv2.imread(filename_x)
    img_x=img_x[:,:,::-1]
    imgArr_x=crop_image(img_x)

#     get file number
    str_f=filename_x.split('optixPathTracernoise_')[1]
    num=int(str_f.split('.ppm')[0])

#     get normal
    filename_n=(path_n+repr(num)+".ppm")
    img_n=cv2.imread(filename_n)
    img_n=img_n[:,:,::-1]
    imgArr_n=crop_image(img_n)

#     get depth
    filename_d=(path_d+repr(num)+".ppm")
    img_d=cv2.imread(filename_d)
    img_d=img_d[:,:,::-1]
    imgArr_d=crop_image(img_d)

#     get ground truth
    filename_y=(path_o+repr(num)+".ppm")
    img_y=cv2.imread(filename_y)
    img_y=img_y[:,:,::-1]
    imgArr_y=crop_image(img_y)


    training_img=merge_img(imgArr_x,imgArr_y,imgArr_n,imgArr_d,i)
    
```
